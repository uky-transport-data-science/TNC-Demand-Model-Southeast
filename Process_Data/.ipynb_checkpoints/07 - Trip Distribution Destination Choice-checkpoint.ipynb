```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('E:/CR2/Repos/TNC-Demand-Model-Southeast/Cleaned_Inputs/Intermediate/utiltiy_and_probability_output.csv')
```

```python
df = df[['ORIGIN', 'DESTINATION', 'DISTANCE', 'LOGSUM']]
```

```python
dropoffs = pd.read_csv('E:/CR2/Repos/TNC-Demand-Model-Southeast/Cleaned_Inputs/Intermediate/fayette_estimated_dropoffs.csv')
```

```python
df = df.merge(dropoffs, on = ['DESTINATION'])
```

```python
# we don't have any tourist census tracts
df['TOURIST_FLAGGER'] = 0
```

```python
df['AIRPORT_FLAGGER'] = np.where(df['DESTINATION'] == 21067004207, 1, 0)
```

```python
df['INTERNAL_FLAGGER'] = np.where(df['ORIGIN'] == df['DESTINATION'], 1, 0)
```

```python
df['NT_UTILITY'] = np.log(df['NT_PRED_DROPOFFS']) - 0.07*df['INTERNAL_FLAGGER'] + 0.49*df['TOURIST_FLAGGER'] - 2.49*df['AIRPORT_FLAGGER'] + 0.85*df['LOGSUM']
df['AM_UTILITY'] = np.log(df['AM_PRED_DROPOFFS']) - 0.07*df['INTERNAL_FLAGGER'] + 0.49*df['TOURIST_FLAGGER'] - 2.49*df['AIRPORT_FLAGGER'] + 0.85*df['LOGSUM']
df['MD_UTILITY'] = np.log(df['MD_PRED_DROPOFFS']) - 0.07*df['INTERNAL_FLAGGER'] + 0.49*df['TOURIST_FLAGGER'] - 2.49*df['AIRPORT_FLAGGER'] + 0.85*df['LOGSUM']
df['PM_UTILITY'] = np.log(df['PM_PRED_DROPOFFS']) - 0.07*df['INTERNAL_FLAGGER'] + 0.49*df['TOURIST_FLAGGER'] - 2.49*df['AIRPORT_FLAGGER'] + 0.85*df['LOGSUM']
df['EV_UTILITY'] = np.log(df['EV_PRED_DROPOFFS']) - 0.07*df['INTERNAL_FLAGGER'] + 0.49*df['TOURIST_FLAGGER'] - 2.49*df['AIRPORT_FLAGGER'] + 0.85*df['LOGSUM']
```
