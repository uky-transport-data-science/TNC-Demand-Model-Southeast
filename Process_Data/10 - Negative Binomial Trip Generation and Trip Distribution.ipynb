```python
# Import Libraries
import numpy as np
import pandas as pd
import os
```

```python
## Read in Data
pickups = pd.read_csv('Cleaned_Inputs/Intermediate/fayette_linear_pickups.csv')
pickups.head()
```

```python
dest_choice = pd.read_csv('Cleaned_Inputs/Intermediate/dest_choice_prob_logsum.csv')
dest_choice.head()
```

```python
df = pd.merge(pickups, dest_choice, left_on = ['origin'], right_on = ['origin'])
df.head()
```

```python
df = df[['origin', 'destination', 'dest_choice_logsum', 'NT_PRED_PICKUPS_LOG', 'AM_PRED_PICKUPS_LOG', 'MD_PRED_PICKUPS_LOG','PM_PRED_PICKUPS_LOG', 'EV_PRED_PICKUPS_LOG', 'PRED_PICKUPS_LOG', 'airport', 'tourist', 'MEDIAN_AGE', 'P_BACH_25P', 'TOTAL_EMP_DEN', 'NT_PROB', 'AM_PROB', 'MD_PROB', 'PM_PROB', 'EV_PROB']]
df.head()
```

```python
## Apply Negative Binomial Model to get pickups
df['NT_PICKUPS'] = np.exp(-1.699553*df['airport'] + 0.5047479*df['tourist'] + 0.3486627*df['NT_PRED_PICKUPS_LOG'] + 1.082214*df['dest_choice_logsum'] - 0.0173114*df['MEDIAN_AGE'] + 0.0170123*df['P_BACH_25P'] + 0.000000716*df['TOTAL_EMP_DEN'] + 0.2991526 - 5.691818)
df['AM_PICKUPS'] = np.exp(-0.7743786*df['airport'] + 0.4779832*df['tourist'] + 0.4050372*df['AM_PRED_PICKUPS_LOG'] + 0.6230226*df['dest_choice_logsum'] - 0.0262*df['MEDIAN_AGE'] + 0.0143772*df['P_BACH_25P']+ 0.000000498*df['TOTAL_EMP_DEN'] + 0.3775664 - 1.946639)
df['MD_PICKUPS'] = np.exp(-0.6405087*df['airport'] + 0.7809523*df['tourist'] + 0.3927277*df['MD_PRED_PICKUPS_LOG'] + 0.6977969*df['dest_choice_logsum'] - 0.0208753*df['MEDIAN_AGE'] + 0.0123951*df['P_BACH_25P']+ 0.00000161*df['TOTAL_EMP_DEN'] + 0.3918755 - 2.551229)
df['PM_PICKUPS'] = np.exp(-0.4855696*df['airport'] + 0.890186*df['tourist'] + 0.4090155*df['PM_PRED_PICKUPS_LOG'] + 0.8352944*df['dest_choice_logsum'] - 0.0216488*df['MEDIAN_AGE'] + 0.0149257*df['P_BACH_25P']+ 0.00000217*df['TOTAL_EMP_DEN'] + 0.363544 - 4.089885)
df['EV_PICKUPS'] = np.exp(0.0188928*df['airport'] + 0.830838*df['tourist'] + 0.3445732*df['EV_PRED_PICKUPS_LOG'] + 0.9817033*df['dest_choice_logsum'] - 0.0215845*df['MEDIAN_AGE'] + 0.0199614*df['P_BACH_25P']+ 0.00000224*df['TOTAL_EMP_DEN'] + 0.4076045 - 5.265743)
df.head()
```

```python
#### Apply Probabilities
time_of_day = ["NT", "AM", "MD", "PM", "EV"]
for time in time_of_day:
    df[time + "_TRIPS"] = df[time + "_PICKUPS"] * df[time + "_PROB"]
df.head()
```

```python
neg_bin_pred_trips = df[['origin', 'destination', 'NT_TRIPS', 'AM_TRIPS', 'MD_TRIPS', 'PM_TRIPS', 'EV_TRIPS']]
neg_bin_pred_trips.to_csv('Cleaned_Inputs/Intermediate/neg_bin_pred_trips.csv', index = False)
```
