```python
# calculating trip fares
```

```python
# private trip fares using 1.9035 + (0.2892)(travel time in minutes) + (0.8822)(distance in miles)
```

```python
trips.loc[:, 'NT_PRIVATE_TRIP_FARES'] = 1.9035 + (0.2892 * trips['NT_PRIVATE_TRAVEL_TIME']) + (0.8822 * trips['DISTANCE_IN_MILES'])
trips.loc[:, 'AM_PRIVATE_TRIP_FARES'] = 1.9035 + (0.2892 * trips['AM_PRIVATE_TRAVEL_TIME']) + (0.8822 * trips['DISTANCE_IN_MILES'])
trips.loc[:, 'MD_PRIVATE_TRIP_FARES'] = 1.9035 + (0.2892 * trips['MD_PRIVATE_TRAVEL_TIME']) + (0.8822 * trips['DISTANCE_IN_MILES'])
trips.loc[:, 'PM_PRIVATE_TRIP_FARES'] = 1.9035 + (0.2892 * trips['PM_PRIVATE_TRAVEL_TIME']) + (0.8822 * trips['DISTANCE_IN_MILES'])
trips.loc[:, 'EV_PRIVATE_TRIP_FARES'] = 1.9035 + (0.2892 * trips['EV_PRIVATE_TRAVEL_TIME']) + (0.8822 * trips['DISTANCE_IN_MILES'])
```

```python
# shared trip fares using 2.6496 + (0.06)(travel time in minutes) + (0.7349)(distance in miles)
```

```python
trips.loc[:, 'NT_SHARED_TRIP_FARES'] = 2.6496 + (0.06 * trips['NT_SHARED_TRAVEL_TIME']) + (0.7349 * trips['DISTANCE_IN_MILES'])
trips.loc[:, 'AM_SHARED_TRIP_FARES'] = 2.6496 + (0.06 * trips['AM_SHARED_TRAVEL_TIME']) + (0.7349* trips['DISTANCE_IN_MILES'])
trips.loc[:, 'MD_SHARED_TRIP_FARES'] = 2.6496 + (0.06 * trips['MD_SHARED_TRAVEL_TIME']) + (0.7349 * trips['DISTANCE_IN_MILES'])
trips.loc[:, 'PM_SHARED_TRIP_FARES'] = 2.6496 + (0.06 * trips['PM_SHARED_TRAVEL_TIME']) + (0.7349 * trips['DISTANCE_IN_MILES'])
trips.loc[:, 'EV_SHARED_TRIP_FARES'] = 2.6496 + (0.06 * trips['EV_SHARED_TRAVEL_TIME']) + (0.7349 * trips['DISTANCE_IN_MILES'])
```
