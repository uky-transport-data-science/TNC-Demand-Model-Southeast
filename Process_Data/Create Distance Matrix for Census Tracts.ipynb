```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
```

```python
points = pd.read_csv('E:/CR2/Repos/TNC-Demand-Model-Southeast/Cleaned_Inputs/Intermediate/centroid_points.csv')
```

```python
points.head()
```

```python
tracts = points.GEOID
```

```python
def find_distance(row, origin):
    dest_lat = row['Y']
    dest_lon = row['X']
    dest = (dest_lat, dest_lon)

    dist = geodesic(origin, dest).mi

    return dist
```

```python
df = pd.DataFrame()
for tract in tracts:
    print(tract)
    origin_lat = points[points['GEOID'] == tract]['Y'].mean()
    origin_lon = points[points['GEOID'] == tract]['X'].mean()
    origin = (origin_lat, origin_lon)
    points['DISTANCE_IN_MILES'] = points.apply(lambda row: find_distance(row, origin), axis = 1)
    points['ORIGIN'] = tract
    df = pd.concat([df, points])
    

# reset index without adding the old index
df.reset_index(drop=True, inplace=True)
```

```python
df['DESTINATION'] = df['GEOID']
```

```python
df = df[['ORIGIN', 'DESTINATION', 'DISTANCE_IN_MILES']]
```
